## 1. Setup

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

## 2. Line Plot

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin(x)", color="blue")
plt.plot(x, y2, label="cos(x)", color="red")
plt.title("Line Plot Example")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

## 3. Scatter Plot

```python
x1 = np.random.rand(50)
y1 = np.random.rand(50)
sizes1 = 300 * np.random.rand(50)
colors1 = np.random.rand(50)

x2 = np.random.rand(50)
y2 = np.random.rand(50)
sizes2 = 300 * np.random.rand(50)
colors2 = np.random.rand(50)

plt.scatter(x1, y1, s=sizes1, c=colors1, alpha=0.6, cmap='viridis')
plt.scatter(x2, y2, s=sizes2, c=colors2, alpha=0.6, cmap='plasma')
plt.title("Scatter Plot Example")
plt.xlabel("x")
plt.ylabel("y")
plt.colorbar(label="color scale")
plt.show()
```

## 4. Bar Plot

```python
categories = ["A", "B", "C", "D"]
values = [5, 7, 3, 8]

plt.bar(categories, values)
plt.title("Bar Plot Example")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
```

## 5. Customizing Plots

```python
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin(x)", linestyle="--", color="blue")
plt.plot(x, y2, label="cos(x)", linestyle=":", color="red")
plt.title("Customized Line Plot")
plt.xlabel("x (radians)")
plt.ylabel("value")
plt.legend()
plt.grid(True)
plt.show()
```

âœ… **Your Turn**: Customize your own plot with labels, legend, and grid lines.

```python
x = np.linspace(-2, 2, 100)
y1 = x**2
y2 = x**3

plt.plot(x, y1, "g--", linewidth=2, label="y = x^2")
plt.plot(x, y2, "m:", linewidth=2, label="y = x^3")
plt.title("Quadratic vs Cubic")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()
```

### Summary
- `%matplotlib inline` makes plots display inside the notebook.
- Line, scatter, and bar plots are the most common basics.
- Titles, labels, legends, and grids improve readability.
- Visuals tell stories about data more effectively than raw numbers.
