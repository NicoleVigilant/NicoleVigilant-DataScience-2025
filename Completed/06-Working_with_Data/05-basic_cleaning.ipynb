# ðŸ§¼ Cleaning Dirty Data (Missing Values & Type Fixes)

## ðŸ”¹ LEARNING GOALS:
- Detect and count missing values (`NaN`)
- Fill or drop missing data
- Convert column data types safely
- Understand the difference between `NaN`, `None`, `""`, and type mismatches


### ðŸ§ª 1. Load a Messy Dataset

```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Alice", "Bob", "Charlie", "David", None],
    "Age": ["25", "thirty", 35, np.nan, "40"],
    "Signup Date": ["2022-01-01", "not a date", "2022/03/01", None, "April 5, 2022"],
    "Score": [95.5, None, 88.0, 92.5, ""]
}

df = pd.DataFrame(data)
df
```

### ðŸ§¯ 2. Detecting Missing or Broken Values

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

### ðŸ§¹ 3. Cleaning Strategy Options

```python
df.fillna({
    "Name": "Unknown",
    "Age": -1,
    "Signup Date": "1970-01-01",
    "Score": 0.0
})
```

```python
df.dropna()
```

### ðŸ§¬ 4. Data Type Fixes

```python
df.dtypes
```

```python
df["Age"] = pd.to_numeric(df["Age"], errors="coerce")
df["Score"] = pd.to_numeric(df["Score"], errors="coerce")
df["Signup Date"] = pd.to_datetime(df["Signup Date"], errors="coerce")
df.dtypes
```

### ðŸ©¹ 5. Impute (Fill In) Fixed Missing Values

```python
df["Age"].fillna(df["Age"].median(), inplace=True)
df["Score"].fillna(df["Score"].mean(), inplace=True)
```

```python
df["Signup Date"].fillna(df["Signup Date"].min(), inplace=True)
```

```python
df["Name"].fillna("Unknown", inplace=True)
```

### ðŸ¤“ 6. Cleaned Data Review

```python
df.info()
```

```python
df.describe(include="all")
```

```python
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", None, "Eve"],
    "Age": ["25", "thirty", 35, np.nan, "40", "twenty-two"],
    "Signup Date": ["2022-01-01", "not a date", "2022/03/01", None, "April 5, 2022", "13/13/2022"],
    "Score": [95.5, None, 88.0, 92.5, "", "eighty"],
    "Email": ["alice@email.com", "", None, "david@email.com", "no email", "eve[at]email"]
}

df = pd.DataFrame(data)
df
```

```python
# Detecting missing or broken values
df.isnull()
```

```python
# Counting nulls before cleaning
df.isnull().sum()
```

```python
# Rows that have missing or broken values
df[df.isnull().any(axis=1)]
```

```python
# Convert "" to np.nan
df = df.replace("", np.nan)
```

```python
# Check types before conversion
df.dtypes
```

```python
# Convert to proper types
df["Age"] = pd.to_numeric(df["Age"], errors="coerce")
df["Score"] = pd.to_numeric(df["Score"], errors="coerce")
df["Signup Date"] = pd.to_datetime(df["Signup Date"], errors="coerce")

# Show data types after conversion
df.dtypes
```

```python
#Filling in missing values
median_age = df["Age"].median()     # middle age (ignores NaN)
mean_score = df["Score"].mean()     # average score (ignores NaN)
date = df["Signup Date"].min()      # earliest valid signup date

print("Age:", median_age)
print("Score:", mean_score)
print("Signup Date:", date)

df["Age"].fillna(median_age, inplace=True)
df["Score"].fillna(mean_score, inplace=True)
df["Signup Date"].fillna(date, inplace=True)
```

```python
# Impute missing values
df["Email"] = df["Email"].replace(["", "no email", "eve[at]email"], np.nan)

df = df.fillna({
    "Name": "Unknown",
    "Age": median_age,
    "Score": mean_score,
    "Signup Date": date,
    "Email": "example@gmail.com"
})

# Verify the result
print(df.isnull().sum())
```

```python
#Verify that everything is clean

df.info()
```

```python
# Print DataFrame with fixed values
print(df)
```

```python
df.describe(include="all")
```
