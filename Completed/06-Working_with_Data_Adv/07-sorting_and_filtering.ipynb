# ðŸ§® Sorting & Filtering in Pandas

## ðŸ”¹ LEARNING GOALS:
- Sort rows using `.sort_values()`
- Filter data using conditional logic
- Chain multiple filters together
- Use `.query()` for readable filtering


### ðŸ“¥ 1. Load the Dataset

```python
# Load the "students.csv" file found in data in assignments folder
from google.colab import files
uploaded = files.upload()

import pandas as pd

df = pd.read_csv("students.csv")
df.head()
```

### ðŸ”¢ 2. Sorting Data

```python
# Sort by math_score
df.sort_values(by="math_score")
```

```python
# Sort by science_score descending
df.sort_values(by="science_score", ascending=False)
```

```python
# Sort by multiple columns
df.sort_values(by=["science_score", "math_score"], ascending=[False, True])
```

### ðŸ§ª 3. Filtering with Conditions

```python
# Who scored at least 90 in math?
df[df["math_score"] >= 90]
```

```python
# Who scored less than 70 in either subject?
df[(df["math_score"] < 70) | (df["science_score"] < 70)]
```

### ðŸ”— 4. Combining Filters

```python
# Students with B+ or higher in both
df[(df["math_score"] >= 87) & (df["science_score"] >= 87)]
```

### ðŸ§¾ 5. Using `.query()` for Readability

```python
# Same query using .query()
df.query("math_score >= 87 and science_score >= 87")
```

### ðŸ§ª Try It Yourself

- Sort the dataset by first name alphabetically
- Find students with the same math and science score
- Use `.query()` to find students who scored over 95 in *either* subject


```python
# Sort by first name alphabetically
df.sort_values(by="first_name", ascending=True)
```

```python
# Find students with same Math and Science score
df[df["math_score"] == df["science_score"]]
```

```python
# Use .query() to find students scoring over 95 in either subject
df.query("math_score > 95 or science_score > 95")
```

I am using the mock_survey.csv that I made since I couldnt find survey.csv

```python

# Load the "mock_survey.csv" file
from google.colab import files
uploaded = files.upload()
```

```python
df= pd.read_csv("mock_survey.csv")
df.head()
```

```python
# Sort by the most recent signup date
df["Signup Date"] = pd.to_datetime(df["Signup Date"], errors="coerce")
df = df.sort_values("Signup Date", ascending=False)
```

```python
# Filter out anyone who didn't complete the survey (nulls or "N/A")
df = df[df["Completed"].notna()]
df = df[df["Completed"] != "N/A"]
```

```python
# Filter out anyone who didn't complete the survey
df.dropna()
```

```python
# Use .query() to select rows where the survey was completed
df.query("Completed == 'Yes'")
```
